```python
#villa_rent 통합
import pandas as pd
import os

# --- 설정 ---
# 작업할 파일 이름 목록
file_names = [
    'villa_rent_2022.csv',
    'villa_rent_2023.csv',
    'villa_rent_2024.csv'
]
# 저장할 최종 파일 이름
output_filename = 'villa_rent_merged_with_type.csv'
# ----------------

# --- 파일 경로 설정 ---
# 이 스크립트가 실행되는 폴더를 기준으로 작업
try:
    script_dir = os.path.dirname(os.path.abspath(__file__))
    os.chdir(script_dir)
except NameError:
    print("현재 작업 폴더에서 파일을 찾습니다...")
    pass
# --------------------

df_list = []
print("--- 파일 읽기 시작 ---")
all_files_found = True
for file in file_names:
    if not os.path.exists(file):
        print(f"❌ [실패] '{file}' 파일을 현재 폴더에서 찾을 수 없습니다! 파일 이름과 위치를 확인하세요.")
        all_files_found = False
        break
    
    try:
        # 국토부 원본 데이터는 보통 cp949 인코딩을 사용한다.
        df = pd.read_csv(file, encoding='cp949', skiprows=15)
        df_list.append(df)
        print(f"✔️ [성공] '{file}' 파일 읽기 성공.")
    except Exception as e:
        print(f"❌ [실패] '{file}' 파일을 읽는 중 오류 발생: {e}")
        all_files_found = False
        break

if all_files_found and df_list:
    # --- 1. 데이터 통합 ---
    # 리스트에 담긴 모든 데이터프레임을 하나로 합친다.
    merged_df = pd.concat(df_list, ignore_index=True)
    print("\n✔️ 3개년 데이터 통합 완료.")

    # --- 2. '주택유형' 피쳐 추가 ---
    # '주택유형'이라는 새 열을 만들고, 모든 값을 '연립다세대'로 채운다.
    merged_df['주택유형'] = '연립다세대'
    print("✔️ '주택유형' 열 추가 완료: 모든 값에 '연립다세대'를 할당했습니다.")

    # --- 3. 결과 저장 ---
    merged_df.to_csv(output_filename, index=False, encoding='utf-8-sig')
    
    print("\n--------------------------------------------------")
    print(f"🎉 모든 작업 완료! '{output_filename}' 파일이 생성되었습니다.")
    print("최종 데이터셋 정보:")
    merged_df.info()
    print("--------------------------------------------------")
else:
    print("\n작업에 실패했습니다. 위의 에러 메시지를 확인하세요.")
```

```python
#rent에서 필요없는 피쳐 지우기 -> rent1
import pandas as pd
import os

# --- 설정 ---
input_filename = 'rent.csv'
output_filename = 'rent1.csv'
# ----------------

# --- 파일 경로 설정 ---
try:
    script_dir = os.path.dirname(os.path.abspath(__file__))
    os.chdir(script_dir)
except NameError:
    print("현재 작업 폴더에서 파일을 찾습니다...")
    pass
# --------------------

try:
    if not os.path.exists(input_filename):
        raise FileNotFoundError(f"'{input_filename}' 파일을 현재 폴더에서 찾을 수 없습니다.")

    # 최종 통합본인 rent.csv 파일을 불러온다.
    df = pd.read_csv(input_filename)
    print(f"✔️ [성공] '{input_filename}' 파일 읽기 성공.")
    print("\n--- 삭제 전 열 목록 ---")
    print(df.columns.tolist())

    # --- 삭제할 열 목록 ---
    # 네가 요청한 열 이름 리스트다.
    columns_to_drop = [
        '번지', '본번', '부번', '전월세구분', '전용면적(㎡)', '월세금(만원)',
        '층', '게약기간', # '계약기간'의 오타일 수 있으나 일단 포함
        '계약기간', # 정확한 이름도 포함
        '계약구분', '갱신요구권 사용', '종전계약 보증금(만원)', '종전계약 월세(만원)',
        '건물명', # 원본 데이터의 '단지명'과 다를 수 있어 일단 포함
        '단지명' # 실제 이름도 포함
    ]
    
    # 데이터프레임에 실제로 존재하는 열만 추려낸다.
    existing_columns_to_drop = [col for col in columns_to_drop if col in df.columns]
    
    if not existing_columns_to_drop:
        print("\n⚠️ 삭제할 열을 찾을 수 없습니다. 열 이름을 확인해주세요.")
    else:
        # --- 열 삭제 ---
        df.drop(columns=existing_columns_to_drop, inplace=True)
        print(f"\n✔️ 다음 열들을 삭제했습니다: {existing_columns_to_drop}")

    # --- 결과 저장 ---
    df.to_csv(output_filename, index=False, encoding='utf-8-sig')
    
    print("\n--------------------------------------------------")
    print(f"🎉 모든 작업 완료! '{output_filename}' 파일이 생성되었습니다.")
    print("--- 삭제 후 최종 데이터 정보 ---")
    df.info()
    print("--------------------------------------------------")

except FileNotFoundError:
    print(f"\n❌ [에러] '{input_filename}'을 찾을 수 없습니다! 파일 이름과 위치를 확인하세요.")
except Exception as e:
    print(f"\n❌ [에러] 작업 중 예상치 못한 오류 발생: {e}")
```

```python
#건축연한 만들기&시군구를 시군으로 바꾸기 -> rent2.csv
import pandas as pd
import os

# --- 설정 ---
input_filename = 'rent1.csv'
output_filename = 'rent2.csv'
# ----------------

# --- 파일 경로 설정 ---
try:
    script_dir = os.path.dirname(os.path.abspath(__file__))
    os.chdir(script_dir)
except NameError:
    print("현재 작업 폴더에서 파일을 찾습니다...")
    pass
# --------------------

try:
    if not os.path.exists(input_filename):
        raise FileNotFoundError(f"'{input_filename}' 파일을 현재 폴더에서 찾을 수 없습니다.")

    df = pd.read_csv(input_filename)
    print(f"✔️ [성공] '{input_filename}' 파일 읽기 성공.")
    print("\n--- 작업 전 데이터 정보 ---")
    df.info()

    # --- 1. '건축연한' 피쳐 생성 ---
    # 계약년월(YYYYMM)에서 연도만 추출하기 위해 100으로 나눈 몫을 사용
    contract_year = df['계약년월'] // 100
    
    # 건축연한 = 계약연도 - 건축년도
    df['건축연한'] = contract_year - df['건축년도']
    print("\n✔️ '건축연한' 열 추가 완료.")

    # --- 2. '시군' 피쳐 생성 및 기존 '시군구' 열 삭제 ---
    # '시군구' 열의 주소를 공백으로 나눈 후, 앞의 두 부분만 선택하여 다시 공백으로 합친다.
    # 예: '서울특별시 강서구 등촌동' -> ['서울특별시', '강서구', '등촌동'] -> ['서울특별시', '강서구'] -> '서울특별시 강서구'
    df['시군'] = df['시군구'].str.split().str[:2].str.join(' ')
    
    # 기존 '시군구' 열은 삭제한다.
    df.drop(columns=['시군구'], inplace=True)
    print("✔️ '시군' 열 생성 및 기존 '시군구' 열 삭제 완료.")

    # --- 결과 저장 ---
    df.to_csv(output_filename, index=False, encoding='utf-8-sig')
    
    print("\n--------------------------------------------------")
    print(f"🎉 모든 작업 완료! '{output_filename}' 파일이 생성되었습니다.")
    print("--- 작업 후 최종 데이터 정보 ---")
    df.info()
    print("\n--- 작업 후 데이터 샘플 (상위 5개) ---")
    print(df.head())
    print("--------------------------------------------------")

except FileNotFoundError:
    print(f"\n❌ [에러] '{input_filename}'을 찾을 수 없습니다! 파일 이름과 위치를 확인하세요.")
except KeyError as e:
    print(f"\n❌ [에러] 열 이름({e})을 찾을 수 없습니다! 'rent1.csv' 파일에 필요한 열이 있는지 확인하세요.")
except Exception as e:
    print(f"\n❌ [에러] 작업 중 예상치 못한 오류 발생: {e}")
```

```python
#도로명주소로 통합 -> rent3.csv
import pandas as pd
import os

# --- 설정 ---
input_filename = 'rent2.csv'
output_filename = 'rent3.csv'
# ----------------

# --- 파일 경로 설정 ---
try:
    script_dir = os.path.dirname(os.path.abspath(__file__))
    os.chdir(script_dir)
except NameError:
    print("현재 작업 폴더에서 파일을 찾습니다...")
    pass
# --------------------

try:
    if not os.path.exists(input_filename):
        raise FileNotFoundError(f"'{input_filename}' 파일을 현재 폴더에서 찾을 수 없습니다.")

    df = pd.read_csv(input_filename)
    print(f"✔️ [성공] '{input_filename}' 파일 읽기 성공.")

    # --- 1. '도로명주소' 열 생성 ---
    # '시군' 열과 '도로명' 열을 공백 하나를 두고 합친다.
    df['도로명주소'] = df['시군'].astype(str) + ' ' + df['도로명'].astype(str)
    print("\n✔️ '도로명주소' 열 생성 완료.")

    # --- 2. 기존 주소 열 삭제 ---
    # 이제 쓸모없어진 '시군'과 '도로명' 열을 삭제한다.
    df.drop(columns=['시군', '도로명'], inplace=True)
    print("✔️ '시군', '도로명' 열 삭제 완료.")
    
    # --- 결과 저장 ---
    df.to_csv(output_filename, index=False, encoding='utf-8-sig')
    
    print("\n--------------------------------------------------")
    print(f"🎉 모든 작업 완료! '{output_filename}' 파일이 생성되었습니다.")
    print("--- 작업 후 최종 데이터 정보 ---")
    df.info()
    print("\n--- 작업 후 데이터 샘플 (상위 5개) ---")
    print(df.head())
    print("--------------------------------------------------")

except FileNotFoundError:
    print(f"\n❌ [에러] '{input_filename}'을 찾을 수 없습니다! 파일 이름과 위치를 확인하세요.")
except KeyError as e:
    print(f"\n❌ [에러] 열 이름({e})을 찾을 수 없습니다! 'rent2.csv' 파일에 필요한 열이 있는지 확인하세요.")
except Exception as e:
    print(f"\n❌ [에러] 작업 중 예상치 못한 오류 발생: {e}")
```

```python
#매매가 정리 및 평균 계산
import pandas as pd
import os

# --- 설정 ---
input_filename = 'sell.csv'
output_filename = 'sell_cleaned.csv'
# ----------------

# --- 파일 경로 설정 ---
try:
    script_dir = os.path.dirname(os.path.abspath(__file__))
    os.chdir(script_dir)
except NameError:
    print("현재 작업 폴더에서 파일을 찾습니다...")
    pass
# --------------------

try:
    if not os.path.exists(input_filename):
        raise FileNotFoundError(f"'{input_filename}' 파일을 현재 폴더에서 찾을 수 없습니다.")

    df = pd.read_csv(input_filename)
    print(f"✔️ [성공] '{input_filename}' 파일 읽기 성공.")

    # --- 1. 불필요한 열 삭제 ---
    columns_to_drop = ['기준연도', '기준월', '법정동코드']
    existing_columns_to_drop = [col for col in columns_to_drop if col in df.columns]
    df.drop(columns=existing_columns_to_drop, inplace=True, errors='ignore')
    print(f"\n✔️ '{', '.join(existing_columns_to_drop)}' 열 삭제 완료.")

    # --- 2. '도로명주소' 기준 평균 계산 ---
    # '도로명주소'가 같은 행들을 하나로 묶고, 나머지 숫자 열들의 평균을 계산한다.
    # as_index=False 옵션은 '도로명주소'를 일반 열로 유지시켜준다.
    grouped_df = df.groupby('도로명주소', as_index=False).mean()
    print("✔️ '도로명주소' 기준 평균 계산 완료.")
    
    # --- 결과 저장 ---
    grouped_df.to_csv(output_filename, index=False, encoding='utf-8-sig')
    
    print("\n--------------------------------------------------")
    print(f"🎉 모든 작업 완료! '{output_filename}' 파일이 생성되었습니다.")
    print("--- 작업 후 최종 데이터 정보 ---")
    grouped_df.info()
    print("\n--- 작업 후 데이터 샘플 (상위 5개) ---")
    print(grouped_df.head())
    print("--------------------------------------------------")

except FileNotFoundError:
    print(f"\n❌ [에러] '{input_filename}'을 찾을 수 없습니다! 파일 이름과 위치를 확인하세요.")
except Exception as e:
    print(f"\n❌ [에러] 작업 중 예상치 못한 오류 발생: {e}")
```

```python
#전세가, 매매가 데이터 합치기
import pandas as pd
import os

# --- 설정 ---
rent_filename = 'rent3.csv'
sell_filename = 'sell_final.csv'
output_filename = 'data_final.csv'
# ----------------

# --- 파일 경로 설정 ---
try:
    script_dir = os.path.dirname(os.path.abspath(__file__))
    os.chdir(script_dir)
except NameError:
    print("현재 작업 폴더에서 파일을 찾습니다...")
    pass
# --------------------

try:
    # --- 1. 두 파일 읽기 ---
    if not os.path.exists(rent_filename):
        raise FileNotFoundError(f"'{rent_filename}' 파일을 찾을 수 없습니다.")
    if not os.path.exists(sell_filename):
        raise FileNotFoundError(f"'{sell_filename}' 파일을 찾을 수 없습니다.")

    rent_df = pd.read_csv(rent_filename)
    sell_df = pd.read_csv(sell_filename)
    print(f"✔️ [성공] '{rent_filename}'과 '{sell_filename}' 파일 읽기 완료.")
    
    # --- 2. Left Merge 수행 ---
    # left=rent_df: rent3.csv를 기준으로 삼는다.
    # right=sell_df: sell_final.csv의 정보를 가져와 붙인다.
    # on='도로명주소': '도로명주소' 열의 값이 같을 때를 기준으로 한다.
    # how='left': 왼쪽(rent_df)의 모든 데이터는 유지하고, 오른쪽에 짝이 없으면 공란(NaN)으로 둔다.
    final_df = pd.merge(rent_df, sell_df, on='도로명주소', how='left')
    print("\n✔️ Left Merge 작업 완료.")
    
    # --- 3. 결과 저장 ---
    final_df.to_csv(output_filename, index=False, encoding='utf-8-sig')
    
    print("\n--------------------------------------------------")
    print(f"🎉 모든 작업 완료! 최종 데이터 '{output_filename}' 파일이 생성되었습니다.")
    print("--- 최종 데이터 정보 ---")
    final_df.info()
    print("\n--- 최종 데이터 샘플 (상위 5개) ---")
    print(final_df.head())
    print("\n--- 매칭되지 않아 공란이 생긴 데이터 샘플 ---")
    # '매매가' 열이 비어있는(NaN) 데이터만 뽑아서 보여준다.
    print(final_df[final_df['매매가'].isnull()].head())
    print("--------------------------------------------------")

except FileNotFoundError as e:
    print(f"\n❌ [에러] {e}")
except Exception as e:
    print(f"\n❌ [에러] 작업 중 예상치 못한 오류 발생: {e}")
```

```python
#도로명주소 앞으로 보내고 매매가 없는 데이터 지우기 -> data_final2.csv
import pandas as pd
import os

# --- 설정 ---
input_filename = 'data_final.csv'
output_filename = 'data_final2.csv'
# ----------------

# --- 파일 경로 설정 ---
try:
    script_dir = os.path.dirname(os.path.abspath(__file__))
    os.chdir(script_dir)
except NameError:
    print("현재 작업 폴더에서 파일을 찾습니다...")
    pass
# --------------------

try:
    if not os.path.exists(input_filename):
        raise FileNotFoundError(f"'{input_filename}' 파일을 현재 폴더에서 찾을 수 없습니다.")

    df = pd.read_csv(input_filename)
    print(f"✔️ [성공] '{input_filename}' 파일 읽기 완료. (총 {len(df)}개 데이터)")

    # --- 1. '도로명주소'를 첫 번째 열로 옮기기 ---
    # 기존 열 목록을 가져온다.
    cols = df.columns.tolist()
    # 목록에서 '도로명주소'를 맨 앞으로 옮긴다.
    cols.insert(0, cols.pop(cols.index('도로명주소')))
    # 데이터프레임을 새로운 열 순서로 재정렬한다.
    df = df[cols]
    print("\n✔️ '도로명주소' 열을 맨 앞으로 이동 완료.")

    # --- 2. 공시가격과 매매가가 모두 공란인 데이터 지우기 ---
    # 삭제 전 데이터 개수를 기록
    rows_before_drop = len(df)
    
    # dropna를 사용하여 '공시가격', '매매가' 열이 '모두' 비어있는(NaN) 행을 삭제한다.
    df.dropna(subset=['공시가격', '매매가'], how='all', inplace=True)
    
    # 삭제 후 데이터 개수를 기록
    rows_after_drop = len(df)
    print(f"✔️ 공란 데이터 삭제 완료. ({rows_before_drop - rows_after_drop}개 행 삭제)")
    
    # --- 3. 최종 파일로 저장 ---
    df.to_csv(output_filename, index=False, encoding='utf-8-sig')
    
    print("\n--------------------------------------------------")
    print(f"🎉 모든 작업 완료! 최종 데이터 '{output_filename}' 파일이 생성되었습니다.")
    print("--- 최종 데이터 정보 ---")
    df.info()
    print("\n--- 최종 데이터 샘플 (상위 5개) ---")
    print(df.head())
    print("--------------------------------------------------")

except FileNotFoundError as e:
    print(f"\n❌ [에러] {e}")
except Exception as e:
    print(f"\n❌ [에러] 작업 중 예상치 못한 오류 발생: {e}")
```

```python
#보증금, 공시가격, 매매가 모두 정수로 만들기 -> data_final3.csv
import pandas as pd
import os

# --- 설정 ---
input_filename = 'data_final2.csv'
output_filename = 'data_final3.csv'
# ----------------

# --- 파일 경로 설정 ---
try:
    script_dir = os.path.dirname(os.path.abspath(__file__))
    os.chdir(script_dir)
except NameError:
    print("현재 작업 폴더에서 파일을 찾습니다...")
    pass
# --------------------

try:
    if not os.path.exists(input_filename):
        raise FileNotFoundError(f"'{input_filename}' 파일을 현재 폴더에서 찾을 수 없습니다.")

    df = pd.read_csv(input_filename)
    print(f"✔️ [성공] '{input_filename}' 파일 읽기 완료.")

    # --- 1. '보증금' 열 변환 ---
    # '보증금(만원)' 열이 숫자가 아닐 경우를 대비해 에러 없이 숫자로 변환
    df['보증금(만원)'] = pd.to_numeric(df['보증금(만원)'].astype(str).str.replace(',', ''), errors='coerce')
    
    # '보증금' 열 생성 (원 단위)
    df['보증금'] = df['보증금(만원)'] * 10000
    
    # 기존 '보증금(만원)' 열 삭제
    df.drop(columns=['보증금(만원)'], inplace=True)
    print("\n✔️ '보증금' 열 변환 및 단위 변경 완료.")

    # --- 2. 공시가격, 매매가 정수 변환 ---
    # NaN 값이 있을 수 있으므로, 먼저 0 등으로 채운 후 정수로 변환하거나,
    # pandas의 Nullable Integer Type을 사용한다. 여기서는 후자를 사용.
    df['공시가격'] = df['공시가격'].round(0).astype('Int64')
    df['매매가'] = df['매매가'].round(0).astype('Int64')
    print("✔️ '공시가격', '매매가' 열을 정수로 변환 완료.")
    
    # --- 3. 최종 파일로 저장 ---
    df.to_csv(output_filename, index=False, encoding='utf-8-sig')
    
    print("\n--------------------------------------------------")
    print(f"🎉 모든 작업 완료! 최종 데이터 '{output_filename}' 파일이 생성되었습니다.")
    print("--- 최종 데이터 정보 ---")
    df.info()
    print("\n--- 최종 데이터 샘플 (상위 5개) ---")
    print(df.head())
    print("--------------------------------------------------")

except FileNotFoundError as e:
    print(f"\n❌ [에러] {e}")
except Exception as e:
    print(f"\n❌ [에러] 작업 중 예상치 못한 오류 발생: {e}")
```

```python
#계약날짜 합치고 기준금리 넣기
import pandas as pd
import os

# --- 0. 데이터 준비 ---
# 2025년 8월 기준, 한국은행 기준금리 변경 이력 (정확한 데이터)
rate_data = {
    '적용일자': [
        '2021-08-26', '2021-11-25', '2022-01-14', '2022-04-14', '2022-05-26', 
        '2022-07-13', '2022-08-25', '2022-10-12', '2022-11-24', '2023-01-13', 
        '2024-10-11', '2024-11-28', '2025-02-25', '2025-05-29'
    ],
    '기준금리': [
        0.75, 1.00, 1.25, 1.50, 1.75, 2.25, 2.50, 3.00, 3.25, 3.50, 
        3.25, 3.00, 2.75, 2.50
    ]
}
rate_df = pd.DataFrame(rate_data)
# 날짜를 기준으로 작업해야 하므로, 문자열을 날짜 타입으로 변환한다.
rate_df['적용일자'] = pd.to_datetime(rate_df['적용일자'])


# --- 파일 경로 설정 및 불러오기 ---
input_filename = 'data_final3.csv'
output_filename = 'data_final4.csv'

try:
    # 스크립트가 실행되는 폴더를 기준으로 작업
    script_dir = os.path.dirname(os.path.abspath(__file__))
    os.chdir(script_dir)
except NameError:
    print("현재 작업 폴더에서 파일을 찾습니다...")
    pass

if not os.path.exists(input_filename):
     raise FileNotFoundError(f"'{input_filename}' 파일을 현재 폴더에서 찾을 수 없습니다.")

df = pd.read_csv(input_filename)
print(f"✔️ [성공] '{input_filename}' 파일 읽기 완료.")


# --- 1. 계약일자 생성 및 정렬 ---
# '계약년월'과 '계약일'을 합쳐서 날짜 타입의 '계약날짜' 열을 만든다.
df['계약날짜'] = pd.to_datetime(df['계약년월'].astype(str) + df['계약일'].astype(str).str.zfill(2), format='%Y%m%d')

# 날짜를 기준으로 두 데이터 모두 정렬한다. (merge_asof의 필수 조건)
df = df.sort_values('계약날짜')
rate_df = rate_df.sort_values('적용일자')
print("✔️ 데이터 정렬 완료.")

# --- 2. pd.merge_asof 로 시점 병합 ---
# '계약날짜'를 기준으로, 그 날짜 혹은 그 이전에 발표된 가장 최근의 금리를 찾아 붙인다.
final_df = pd.merge_asof(
    df,
    rate_df,
    left_on='계약날짜',
    right_on='적용일자',
    direction='backward' # 과거 방향으로 찾기
)
print("✔️ 기준금리 병합 완료.")

# --- 3. 불필요한 열 정리 및 결과 저장 ---
final_df.drop(columns=['적용일자'], inplace=True, errors='ignore')
final_df.to_csv(output_filename, index=False, encoding='utf-8-sig')

print("\n--------------------------------------------------")
print(f"🎉 모든 작업 완료! 최종 데이터 '{output_filename}' 파일이 생성되었습니다.")
print("--- 최종 데이터 정보 ---")
final_df.info()
print("\n--- 최종 데이터 샘플 (상위 5개) ---")
# 계약날짜와 기준금리가 잘 붙었는지 확인
print(final_df[['계약날짜', '기준금리']].head())
print("--------------------------------------------------")
```

```python
#매수우위지수 파일 세로로 뒤집기
import pandas as pd
import os

# --- 설정 ---
input_filename = 'buyers_index_seoul.xlsx'
output_filename = 'buyers_index_seoul_cleaned.csv'
# ----------------

# --- 파일 경로 설정 ---
try:
    script_dir = os.path.dirname(os.path.abspath(__file__))
    os.chdir(script_dir)
except NameError:
    print("현재 작업 폴더에서 파일을 찾습니다...")
    pass
# --------------------

try:
    if not os.path.exists(input_filename):
        raise FileNotFoundError(f"'{input_filename}' 파일을 현재 폴더에서 찾을 수 없습니다.")

    # 엑셀 파일을 읽는다
    df = pd.read_excel(input_filename)
    print(f"✔️ [성공] '{input_filename}' 파일 읽기 완료.")

    # --- 1. '매수우위지수' 행만 선택 ---
    # '조사항목' 열 이름이 미세하게 다를 수 있으니, 앞뒤 공백 제거 후 비교
    df['조사항목'] = df['조사항목'].str.strip()
    index_df = df[df['조사항목'] == '매수우위지수'].copy()

    if index_df.empty:
        raise ValueError("'조사항목' 열에서 '매수우위지수'를 찾을 수 없습니다. 엑셀 파일 내용을 확인해주세요.")

    # --- 2. 데이터를 '세로'로 세우기 (Melt) ---
    melted_df = index_df.melt(
        id_vars=['조사항목'],
        var_name='날짜',
        value_name='매수우위지수'
    )
    print("\n✔️ 데이터를 세로 형태로 변환 완료.")

    # --- 3. 데이터 정리 ---
    cleaned_df = melted_df.drop(columns=['조사항목'])
    cleaned_df['날짜'] = pd.to_datetime(cleaned_df['날짜'])
    cleaned_df['계약년월'] = cleaned_df['날짜'].dt.year * 100 + cleaned_df['날짜'].dt.month
    final_index_df = cleaned_df[['계약년월', '매수우위지수']]
    print("✔️ 데이터 정리 및 '계약년월' 키 생성 완료.")
    
    # --- 4. 결과 저장 ---
    final_index_df.to_csv(output_filename, index=False, encoding='utf-8-sig')
    
    print("\n--------------------------------------------------")
    print(f"🎉 모든 작업 완료! '{output_filename}' 파일이 생성되었습니다.")
    print("--- 최종 데이터 샘플 (상위 5개) ---")
    print(final_index_df.head())
    print("--------------------------------------------------")

except FileNotFoundError:
    print(f"\n❌ [에러] '{input_filename}'을 찾을 수 없습니다! 파일 이름과 위치를 확인하세요.")
except ValueError as e:
    print(f"\n❌ [에-러] {e}")
except Exception as e:
    print(f"\n❌ [에러] 작업 중 예상치 못한 오류 발생: {e}")
```

```python
#매수우위지수 추가하기
import pandas as pd
import os

# --- 설정 ---
main_data_filename = 'data_final4.csv'
index_data_filename = 'buyers_index_seoul_cleaned.csv'
output_filename = 'data_final5.csv'
# ----------------

# --- 파일 경로 설정 ---
try:
    script_dir = os.path.dirname(os.path.abspath(__file__))
    os.chdir(script_dir)
except NameError:
    print("현재 작업 폴더에서 파일을 찾습니다...")
    pass
# --------------------

try:
    if not os.path.exists(main_data_filename):
        raise FileNotFoundError(f"'{main_data_filename}' 파일을 찾을 수 없습니다.")
    if not os.path.exists(index_data_filename):
        raise FileNotFoundError(f"'{index_data_filename}' 파일을 찾을 수 없습니다.")

    # 두 파일을 읽는다.
    df = pd.read_csv(main_data_filename)
    index_df = pd.read_csv(index_data_filename)
    print("✔️ 두 파일 모두 읽기 성공.")
    
    # --- Left Merge 수행 ---
    # '계약년월'을 기준으로, 메인 데이터(df)에 매수우위지수 데이터(index_df)를 붙인다.
    final_df = pd.merge(df, index_df, on='계약년월', how='left')
    print("\n✔️ '계약년월' 기준 병합 완료.")
    
    # --- 결과 저장 ---
    final_df.to_csv(output_filename, index=False, encoding='utf-8-sig')
    
    print("\n--------------------------------------------------")
    print(f"🎉 모든 작업 완료! 최종 데이터 '{output_filename}' 파일이 생성되었습니다.")
    print("--- 최종 데이터 정보 ---")
    final_df.info()
    print("\n--- 최종 데이터 샘플 (상위 5개) ---")
    print(final_df[['계약년월', '매수우위지수']].head())
    print("--------------------------------------------------")

except FileNotFoundError as e:
    print(f"\n❌ [에러] {e}")
except Exception as e:
    print(f"\n❌ [에러] 작업 중 예상치 못한 오류 발생: {e}")
```

```python
#전세가율 계산
import pandas as pd
import numpy as np
import os

# --- 설정 ---
input_filename = 'data_final5.csv'
output_filename = 'data_final6.csv'
# ----------------

# --- 파일 경로 설정 ---
try:
    script_dir = os.path.dirname(os.path.abspath(__file__))
    os.chdir(script_dir)
except NameError:
    print("현재 작업 폴더에서 파일을 찾습니다...")
    pass
# --------------------

try:
    if not os.path.exists(input_filename):
        raise FileNotFoundError(f"'{input_filename}' 파일을 현재 폴더에서 찾을 수 없습니다.")

    df = pd.read_csv(input_filename)
    print(f"✔️ [성공] '{input_filename}' 파일 읽기 완료.")

    # --- 전세가율 계산 ---
    # '매매가'가 0인 경우, 0으로 나누면 무한대(inf)가 되어 계산에 오류를 일으킬 수 있다.
    # 이를 방지하기 위해 매매가가 0인 값을 빈 값(NaN)으로 임시 변경한다.
    df['매매가'] = df['매매가'].replace(0, np.nan)
    
    # 전세가율 = 보증금 / 매매가
    df['전세가율'] = df['보증금'] / df['매매가']
    print("\n✔️ '전세가율' 계산 완료.")

    # --- 결과 저장 ---
    df.to_csv(output_filename, index=False, encoding='utf-8-sig')
    
    print("\n--------------------------------------------------")
    print(f"🎉 모든 작업 완료! '{output_filename}' 파일이 생성되었습니다.")
    print("--- 최종 데이터 정보 ---")
    df.info()
    print("\n--- 최종 데이터 샘플 (상위 5개) ---")
    # 주요 계산 열들을 확인
    print(df[['보증금', '매매가', '전세가율']].head())
    print("--------------------------------------------------")

except FileNotFoundError as e:
    print(f"\n❌ [에러] {e}")
except KeyError as e:
    print(f"\n❌ [에러] {e} 열을 찾을 수 없습니다. 'data_final5.csv' 파일에 '보증금'과 '매매가' 열이 있는지 확인해주세요.")
except Exception as e:
    print(f"\n❌ [에러] 작업 중 예상치 못한 오류 발생: {e}")
```

```python
#'전세가율' 데이터 정제, 카테고리 데이터 변환
import pandas as pd
import os

# --- 설정 ---
input_filename = 'data_final6.csv'
output_filename = 'preprocessed_final.csv'
# ----------------

# --- 파일 경로 설정 ---
try:
    script_dir = os.path.dirname(os.path.abspath(__file__))
    os.chdir(script_dir)
except NameError:
    print("현재 작업 폴더에서 파일을 찾습니다...")
    pass
# --------------------

try:
    if not os.path.exists(input_filename):
        raise FileNotFoundError(f"'{input_filename}' 파일을 현재 폴더에서 찾을 수 없습니다.")

    df = pd.read_csv(input_filename)
    print(f"✔️ [성공] '{input_filename}' 파일 읽기 완료. (총 {len(df)}개 데이터)")

    # --- 1. '전세가율' 데이터 정제 ---
    rows_before = len(df)
    
    # 1-1. 전세가율이 비어있는(NaN) 행 삭제
    df.dropna(subset=['전세가율'], inplace=True)
    
    # 1-2. 비정상적인 전세가율(이상치) 데이터 삭제
    # 예: 10% 미만, 150% 초과인 경우. 이 기준은 너희 팀이 논의해서 조절할 수 있다.
    df = df[(df['전세가율'] >= 0.1) & (df['전세가율'] <= 1.5)]
    
    rows_after = len(df)
    print(f"\n✔️ '전세가율' 정제 완료. ({rows_before - rows_after}개 행 삭제)")

    # --- 2. 카테고리 데이터 변환 (원-핫 인코딩) ---
    # '주택유형' 열에 대해 원-핫 인코딩을 수행한다.
    # 컴퓨터가 이해할 수 있도록 '주택유형_오피스텔', '주택유형_연립다세대' 같은 열로 바꿔준다.
    dummies = pd.get_dummies(df['주택유형'], prefix='주택유형')
    
    # 원래 데이터프레임에 새로 만들어진 더미 변수 열들을 합친다.
    df = pd.concat([df, dummies], axis=1)
    
    # 이제 쓸모없어진 원래의 '주택유형' 열은 삭제한다.
    df.drop(columns=['주택유형'], inplace=True)
    print("✔️ '주택유형' 열 원-핫 인코딩 완료.")
    
    # --- 3. 최종 파일로 저장 ---
    df.to_csv(output_filename, index=False, encoding='utf-8-sig')
    
    print("\n--------------------------------------------------")
    print(f"🎉 모든 작업 완료! 모델링 팀에 전달할 '{output_filename}' 파일이 생성되었습니다.")
    print("--- 최종 데이터 정보 ---")
    df.info()
    print("\n--- 최종 데이터 샘플 (상위 5개) ---")
    print(df.head())
    print("--------------------------------------------------")

except FileNotFoundError as e:
    print(f"\n❌ [에러] {e}")
except Exception as e:
    print(f"\n❌ [에러] 작업 중 예상치 못한 오류 발생: {e}")
```

```python
#수정: 계약년월, 계약일 제거
import pandas as pd
import os

# --- 설정 ---
filename = 'preprocessed_final.csv'
# ----------------

# --- 파일 경로 설정 ---
try:
    script_dir = os.path.dirname(os.path.abspath(__file__))
    os.chdir(script_dir)
except NameError:
    print("현재 작업 폴더에서 파일을 찾습니다...")
    pass
# --------------------

try:
    if not os.path.exists(filename):
        raise FileNotFoundError(f"'{filename}' 파일을 현재 폴더에서 찾을 수 없습니다.")

    df = pd.read_csv(filename)
    print(f"✔️ [성공] '{filename}' 파일 읽기 완료.")
    print("\n--- 삭제 전 열 목록 ---")
    print(df.columns.tolist())

    # --- 삭제할 열 목록 ---
    columns_to_drop = ['계약년월', '계약일']
    
    # 데이터프레임에 실제 존재하는 열만 추려낸다.
    existing_columns_to_drop = [col for col in columns_to_drop if col in df.columns]
    
    if not existing_columns_to_drop:
         print("\n⚠️ 삭제할 열을 찾을 수 없습니다. 이미 삭제된 것 같습니다.")
    else:
        # --- 열 삭제 ---
        df.drop(columns=existing_columns_to_drop, inplace=True)
        print(f"\n✔️ 다음 열들을 삭제했습니다: {existing_columns_to_drop}")

    # --- 최종 파일로 저장 (덮어쓰기) ---
    df.to_csv(filename, index=False, encoding='utf-8-sig')
    
    print("\n--------------------------------------------------")
    print(f"🎉 모든 작업 완료! '{filename}' 파일이 최종 수정되었습니다.")
    print("--- 삭제 후 최종 데이터 정보 ---")
    df.info()
    print("--------------------------------------------------")

except FileNotFoundError as e:
    print(f"\n❌ [에러] {e}")
except Exception as e:
    print(f"\n❌ [에러] 작업 중 예상치 못한 오류 발생: {e}")
```
