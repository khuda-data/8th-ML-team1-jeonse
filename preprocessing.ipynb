```python
#ë§¤ìˆ˜ìš°ìœ„ì§€ìˆ˜ íŒŒì¼ ì„¸ë¡œë¡œ ë’¤ì§‘ê¸°
import pandas as pd
import os

# --- ì„¤ì • ---
input_filename = 'buyers_index_seoul.xlsx'
output_filename = 'buyers_index_seoul_cleaned.csv'
# ----------------

# --- íŒŒì¼ ê²½ë¡œ ì„¤ì • ---
try:
    script_dir = os.path.dirname(os.path.abspath(__file__))
    os.chdir(script_dir)
except NameError:
    print("í˜„ì¬ ì‘ì—… í´ë”ì—ì„œ íŒŒì¼ì„ ì°¾ìŠµë‹ˆë‹¤...")
    pass
# --------------------

try:
    if not os.path.exists(input_filename):
        raise FileNotFoundError(f"'{input_filename}' íŒŒì¼ì„ í˜„ì¬ í´ë”ì—ì„œ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")

    # ì—‘ì…€ íŒŒì¼ì„ ì½ëŠ”ë‹¤
    df = pd.read_excel(input_filename)
    print(f"âœ”ï¸ [ì„±ê³µ] '{input_filename}' íŒŒì¼ ì½ê¸° ì™„ë£Œ.")

    # --- 1. 'ë§¤ìˆ˜ìš°ìœ„ì§€ìˆ˜' í–‰ë§Œ ì„ íƒ ---
    # 'ì¡°ì‚¬í•­ëª©' ì—´ ì´ë¦„ì´ ë¯¸ì„¸í•˜ê²Œ ë‹¤ë¥¼ ìˆ˜ ìˆìœ¼ë‹ˆ, ì•ë’¤ ê³µë°± ì œê±° í›„ ë¹„êµ
    df['ì¡°ì‚¬í•­ëª©'] = df['ì¡°ì‚¬í•­ëª©'].str.strip()
    index_df = df[df['ì¡°ì‚¬í•­ëª©'] == 'ë§¤ìˆ˜ìš°ìœ„ì§€ìˆ˜'].copy()

    if index_df.empty:
        raise ValueError("'ì¡°ì‚¬í•­ëª©' ì—´ì—ì„œ 'ë§¤ìˆ˜ìš°ìœ„ì§€ìˆ˜'ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ì—‘ì…€ íŒŒì¼ ë‚´ìš©ì„ í™•ì¸í•´ì£¼ì„¸ìš”.")

    # --- 2. ë°ì´í„°ë¥¼ 'ì„¸ë¡œ'ë¡œ ì„¸ìš°ê¸° (Melt) ---
    melted_df = index_df.melt(
        id_vars=['ì¡°ì‚¬í•­ëª©'],
        var_name='ë‚ ì§œ',
        value_name='ë§¤ìˆ˜ìš°ìœ„ì§€ìˆ˜'
    )
    print("\nâœ”ï¸ ë°ì´í„°ë¥¼ ì„¸ë¡œ í˜•íƒœë¡œ ë³€í™˜ ì™„ë£Œ.")

    # --- 3. ë°ì´í„° ì •ë¦¬ ---
    cleaned_df = melted_df.drop(columns=['ì¡°ì‚¬í•­ëª©'])
    cleaned_df['ë‚ ì§œ'] = pd.to_datetime(cleaned_df['ë‚ ì§œ'])
    cleaned_df['ê³„ì•½ë…„ì›”'] = cleaned_df['ë‚ ì§œ'].dt.year * 100 + cleaned_df['ë‚ ì§œ'].dt.month
    final_index_df = cleaned_df[['ê³„ì•½ë…„ì›”', 'ë§¤ìˆ˜ìš°ìœ„ì§€ìˆ˜']]
    print("âœ”ï¸ ë°ì´í„° ì •ë¦¬ ë° 'ê³„ì•½ë…„ì›”' í‚¤ ìƒì„± ì™„ë£Œ.")
    
    # --- 4. ê²°ê³¼ ì €ì¥ ---
    final_index_df.to_csv(output_filename, index=False, encoding='utf-8-sig')
    
    print("\n--------------------------------------------------")
    print(f"ğŸ‰ ëª¨ë“  ì‘ì—… ì™„ë£Œ! '{output_filename}' íŒŒì¼ì´ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤.")
    print("--- ìµœì¢… ë°ì´í„° ìƒ˜í”Œ (ìƒìœ„ 5ê°œ) ---")
    print(final_index_df.head())
    print("--------------------------------------------------")

except FileNotFoundError:
    print(f"\nâŒ [ì—ëŸ¬] '{input_filename}'ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤! íŒŒì¼ ì´ë¦„ê³¼ ìœ„ì¹˜ë¥¼ í™•ì¸í•˜ì„¸ìš”.")
except ValueError as e:
    print(f"\nâŒ [ì—-ëŸ¬] {e}")
except Exception as e:
    print(f"\nâŒ [ì—ëŸ¬] ì‘ì—… ì¤‘ ì˜ˆìƒì¹˜ ëª»í•œ ì˜¤ë¥˜ ë°œìƒ: {e}")
```

```python
#ë§¤ìˆ˜ìš°ìœ„ì§€ìˆ˜ ì¶”ê°€í•˜ê¸°
import pandas as pd
import os

# --- ì„¤ì • ---
main_data_filename = 'data_final4.csv'
index_data_filename = 'buyers_index_seoul_cleaned.csv'
output_filename = 'data_final5.csv'
# ----------------

# --- íŒŒì¼ ê²½ë¡œ ì„¤ì • ---
try:
    script_dir = os.path.dirname(os.path.abspath(__file__))
    os.chdir(script_dir)
except NameError:
    print("í˜„ì¬ ì‘ì—… í´ë”ì—ì„œ íŒŒì¼ì„ ì°¾ìŠµë‹ˆë‹¤...")
    pass
# --------------------

try:
    if not os.path.exists(main_data_filename):
        raise FileNotFoundError(f"'{main_data_filename}' íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
    if not os.path.exists(index_data_filename):
        raise FileNotFoundError(f"'{index_data_filename}' íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")

    # ë‘ íŒŒì¼ì„ ì½ëŠ”ë‹¤.
    df = pd.read_csv(main_data_filename)
    index_df = pd.read_csv(index_data_filename)
    print("âœ”ï¸ ë‘ íŒŒì¼ ëª¨ë‘ ì½ê¸° ì„±ê³µ.")
    
    # --- Left Merge ìˆ˜í–‰ ---
    # 'ê³„ì•½ë…„ì›”'ì„ ê¸°ì¤€ìœ¼ë¡œ, ë©”ì¸ ë°ì´í„°(df)ì— ë§¤ìˆ˜ìš°ìœ„ì§€ìˆ˜ ë°ì´í„°(index_df)ë¥¼ ë¶™ì¸ë‹¤.
    final_df = pd.merge(df, index_df, on='ê³„ì•½ë…„ì›”', how='left')
    print("\nâœ”ï¸ 'ê³„ì•½ë…„ì›”' ê¸°ì¤€ ë³‘í•© ì™„ë£Œ.")
    
    # --- ê²°ê³¼ ì €ì¥ ---
    final_df.to_csv(output_filename, index=False, encoding='utf-8-sig')
    
    print("\n--------------------------------------------------")
    print(f"ğŸ‰ ëª¨ë“  ì‘ì—… ì™„ë£Œ! ìµœì¢… ë°ì´í„° '{output_filename}' íŒŒì¼ì´ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤.")
    print("--- ìµœì¢… ë°ì´í„° ì •ë³´ ---")
    final_df.info()
    print("\n--- ìµœì¢… ë°ì´í„° ìƒ˜í”Œ (ìƒìœ„ 5ê°œ) ---")
    print(final_df[['ê³„ì•½ë…„ì›”', 'ë§¤ìˆ˜ìš°ìœ„ì§€ìˆ˜']].head())
    print("--------------------------------------------------")

except FileNotFoundError as e:
    print(f"\nâŒ [ì—ëŸ¬] {e}")
except Exception as e:
    print(f"\nâŒ [ì—ëŸ¬] ì‘ì—… ì¤‘ ì˜ˆìƒì¹˜ ëª»í•œ ì˜¤ë¥˜ ë°œìƒ: {e}")
```

```python
#'ì „ì„¸ê°€ìœ¨' ë°ì´í„° ì •ì œ, ì¹´í…Œê³ ë¦¬ ë°ì´í„° ë³€í™˜
import pandas as pd
import os

# --- ì„¤ì • ---
input_filename = 'data_final6.csv'
output_filename = 'preprocessed_final.csv'
# ----------------

# --- íŒŒì¼ ê²½ë¡œ ì„¤ì • ---
try:
    script_dir = os.path.dirname(os.path.abspath(__file__))
    os.chdir(script_dir)
except NameError:
    print("í˜„ì¬ ì‘ì—… í´ë”ì—ì„œ íŒŒì¼ì„ ì°¾ìŠµë‹ˆë‹¤...")
    pass
# --------------------

try:
    if not os.path.exists(input_filename):
        raise FileNotFoundError(f"'{input_filename}' íŒŒì¼ì„ í˜„ì¬ í´ë”ì—ì„œ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")

    df = pd.read_csv(input_filename)
    print(f"âœ”ï¸ [ì„±ê³µ] '{input_filename}' íŒŒì¼ ì½ê¸° ì™„ë£Œ. (ì´ {len(df)}ê°œ ë°ì´í„°)")

    # --- 1. 'ì „ì„¸ê°€ìœ¨' ë°ì´í„° ì •ì œ ---
    rows_before = len(df)
    
    # 1-1. ì „ì„¸ê°€ìœ¨ì´ ë¹„ì–´ìˆëŠ”(NaN) í–‰ ì‚­ì œ
    df.dropna(subset=['ì „ì„¸ê°€ìœ¨'], inplace=True)
    
    # 1-2. ë¹„ì •ìƒì ì¸ ì „ì„¸ê°€ìœ¨(ì´ìƒì¹˜) ë°ì´í„° ì‚­ì œ
    # ì˜ˆ: 10% ë¯¸ë§Œ, 150% ì´ˆê³¼ì¸ ê²½ìš°. ì´ ê¸°ì¤€ì€ ë„ˆí¬ íŒ€ì´ ë…¼ì˜í•´ì„œ ì¡°ì ˆí•  ìˆ˜ ìˆë‹¤.
    df = df[(df['ì „ì„¸ê°€ìœ¨'] >= 0.1) & (df['ì „ì„¸ê°€ìœ¨'] <= 1.5)]
    
    rows_after = len(df)
    print(f"\nâœ”ï¸ 'ì „ì„¸ê°€ìœ¨' ì •ì œ ì™„ë£Œ. ({rows_before - rows_after}ê°œ í–‰ ì‚­ì œ)")

    # --- 2. ì¹´í…Œê³ ë¦¬ ë°ì´í„° ë³€í™˜ (ì›-í•« ì¸ì½”ë”©) ---
    # 'ì£¼íƒìœ í˜•' ì—´ì— ëŒ€í•´ ì›-í•« ì¸ì½”ë”©ì„ ìˆ˜í–‰í•œë‹¤.
    # ì»´í“¨í„°ê°€ ì´í•´í•  ìˆ˜ ìˆë„ë¡ 'ì£¼íƒìœ í˜•_ì˜¤í”¼ìŠ¤í…”', 'ì£¼íƒìœ í˜•_ì—°ë¦½ë‹¤ì„¸ëŒ€' ê°™ì€ ì—´ë¡œ ë°”ê¿”ì¤€ë‹¤.
    dummies = pd.get_dummies(df['ì£¼íƒìœ í˜•'], prefix='ì£¼íƒìœ í˜•')
    
    # ì›ë˜ ë°ì´í„°í”„ë ˆì„ì— ìƒˆë¡œ ë§Œë“¤ì–´ì§„ ë”ë¯¸ ë³€ìˆ˜ ì—´ë“¤ì„ í•©ì¹œë‹¤.
    df = pd.concat([df, dummies], axis=1)
    
    # ì´ì œ ì“¸ëª¨ì—†ì–´ì§„ ì›ë˜ì˜ 'ì£¼íƒìœ í˜•' ì—´ì€ ì‚­ì œí•œë‹¤.
    df.drop(columns=['ì£¼íƒìœ í˜•'], inplace=True)
    print("âœ”ï¸ 'ì£¼íƒìœ í˜•' ì—´ ì›-í•« ì¸ì½”ë”© ì™„ë£Œ.")
    
    # --- 3. ìµœì¢… íŒŒì¼ë¡œ ì €ì¥ ---
    df.to_csv(output_filename, index=False, encoding='utf-8-sig')
    
    print("\n--------------------------------------------------")
    print(f"ğŸ‰ ëª¨ë“  ì‘ì—… ì™„ë£Œ! ëª¨ë¸ë§ íŒ€ì— ì „ë‹¬í•  '{output_filename}' íŒŒì¼ì´ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤.")
    print("--- ìµœì¢… ë°ì´í„° ì •ë³´ ---")
    df.info()
    print("\n--- ìµœì¢… ë°ì´í„° ìƒ˜í”Œ (ìƒìœ„ 5ê°œ) ---")
    print(df.head())
    print("--------------------------------------------------")

except FileNotFoundError as e:
    print(f"\nâŒ [ì—ëŸ¬] {e}")
except Exception as e:
    print(f"\nâŒ [ì—ëŸ¬] ì‘ì—… ì¤‘ ì˜ˆìƒì¹˜ ëª»í•œ ì˜¤ë¥˜ ë°œìƒ: {e}")
```
